```python
import pandas as pd

# (1~2단계) SCADA Vestas 전체 데이터에 대해 전처리 적용
scada_vestas_all = pd.read_csv(train_dir / 'scada_vestas_train.csv')

# --- 처방 시작 ---

# 1. 터무니없는 양수(+) 이상치 제거 (정격 용량 3.3MW의 1.1배를 초과하는 값은 제거)
# ※ wtg01 ~ wtg12 전체 컬럼에 대해 적용해야 합니다. 예시로 wtg01만 작성.
scada_filtered = scada_vestas_all[scada_vestas_all['vestas_wtg01_power_kw10m'] < 3.3 * 1.1 * 1000]

# 2. 존재할 수 없는 음수(-) 이상치 제거 (0보다 작은 값은 제거)
scada_filtered = scada_filtered[scada_filtered['vestas_wtg01_power_kw10m'] >= 0]

# 3. 풍속 및 발전량 결측치(NaN) 제거
# 예보 데이터 합치기 전에 SCADA 자체의 NaN부터 정리합니다.
scada_filtered = scada_filtered.dropna(subset=['vestas_wtg01_ws', 'vestas_wtg01_power_kw10m'])

# --- 처방 끝 ---
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 베스타스 터빈 데이터 가볍게 30,000줄만 읽어오기 (용량 걱정 NO!)
scada_vestas = pd.read_csv(train_dir / 'scada_vestas_train.csv', nrows=30000)

# 2. 1번 터빈(wtg01) 데이터만 떼서 산점도 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=scada_vestas, 
    x='vestas_wtg01_ws',        # 1번 터빈 풍속
    y='vestas_wtg01_power_kw10m', # 1번 터빈 발전량
    alpha=0.3, 
    color='teal'
)

plt.title('Vestas WTG01 - Power Curve')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Power (kW)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 원본 데이터 다시 읽기
scada_vestas = pd.read_csv(train_dir / 'scada_vestas_train.csv', nrows=30000)

# 2. 확실하게 필터링해서 새로운 변수(scada_filtered)에 저장하기
# - 발전량이 0 이상이고, 4000 kW 이하인 정상 범위만 남깁니다.
scada_filtered = scada_vestas[
    (scada_vestas['vestas_wtg01_power_kw10m'] >= 0) & 
    (scada_vestas['vestas_wtg01_power_kw10m'] <= 4000)
]

# 3. 필터링된 데이터로 다시 산점도 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=scada_filtered,          # <-- 여기가 중요합니다! 필터링된 데이터를 넣어야 해요.
    x='vestas_wtg01_ws',        
    y='vestas_wtg01_power_kw10m', 
    alpha=0.3, 
    color='teal'
)

plt.title('Vestas WTG01 - Power Curve (Filtered)')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Power (kW)')
plt.grid(True)
plt.show()
```
